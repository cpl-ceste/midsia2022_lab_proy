```python
import pandas as pd
```

# HOUSE PRICING

```python
df=pd.read_csv('house-prices-advanced-regression-techniques-publicleaderboard.csv')
```

```python
df.head()
```

```python
df['Score'].hist(figsize=(20,10),bins=20)
```

```python
for q in [0.25,0.50, .75, 1]:
  print(f'quantile {q} ->'+str(df['Score'].quantile(q)))
```

```python
df_top=pd.DataFrame(df[df['Score']<df['Score'].quantile(.75)]['Score'])
df_top.hist(figsize=(30,10),bins=20,xlabelsize=18,ylabelsize=18)
```

```python
df_top.boxplot(column=['Score'],figsize=(10,10),fontsize=18)
```
